Using two red tiles as opposite corners, what is the largest area of any rectangle you can make?


```python
test_path = "../test/test09_1.txt"
file_path = "day09.txt"
```

```python
def get_coordinates(filepath: str = test_path) -> list:
    '''
    Gets the coordinates from the file specified and returns them as a list of tuples (x, y)
    '''

    with open(filepath, "r") as file:
        lines = file.readlines()
    coordinates = []
    for line in lines:
        x, y = line.strip().split(",")
        coordinates.append((int(x), int(y)))

    return coordinates
```

```python
get_coordinates()
```

```python
def calc_area(coord1: tuple, coord2: tuple) -> int:
    '''
    Calculates the area spanned by the two coordinates by using the two coordinates as opposing corners of the square. 
    Returns the area of the square as an int.
    '''

    return (abs(coord1[0]-coord2[0])+1)*(abs(coord1[1]-coord2[1])+1)
```

```python
def test_calc_area():
    assert calc_area((2, 5), (9, 7) ) == 24
    assert calc_area((7, 1), (11, 7) ) == 35
    assert calc_area((7, 3), (2, 3) ) == 6
    assert calc_area((2, 5), (11, 1) ) == 50
```

```python
test_calc_area()
```

```python
def day09a(filepath: str = test_path) -> int:
    '''
    
    '''

    coordinates = get_coordinates(filepath)
    
    largest = 0
    for i, coord1 in enumerate(coordinates):
        for j, coord2 in enumerate(coordinates[i+1:]):
            largest = max(largest, calc_area(coord1, coord2))

    print("Day 09 part a:")
    print(f"The largest area that can be made is {largest}.")
    return largest
```

```python
def test_day09a():
    assert day09a() == 50
```

```python
test_day09a()
```
